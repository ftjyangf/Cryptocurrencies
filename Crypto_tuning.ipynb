```python
import pandas as pd
import hvplot.pandas
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.cluster import KMeans
```

# based on last analysis, for visualization propose, 3 components were selected for pca. but it doesn't mean we kept enough infomation in the process in this task, we decide to using different scaler, and pca components to see what kmeans cluster can present us

```python
file_path = 'crypto_data.csv'
crypto = pd.read_csv(file_path)
```

```python
crypto = crypto[crypto['IsTrading'] == True].set_index('Unnamed: 0')
crypto = crypto[crypto['Algorithm'].notnull()]

crypto =crypto.drop('IsTrading',axis = 1)
print(crypto.shape)
crypto = crypto.dropna()
crypto =crypto[crypto['TotalCoinsMined'] > 0]
CoinName = crypto['CoinName']
crypto1 = crypto.drop(columns = 'CoinName')
crypto2 = pd.get_dummies(crypto1,columns = ['Algorithm','ProofType'])
```

```python
# stardardscaler and minscaler
ss_crypto = StandardScaler().fit_transform(crypto2)
mm_crypto = MinMaxScaler().fit_transform(crypto2)
print(ss_crypto)
print(mm_crypto)
```

```python
# pca explained_variance ratio 
%matplotlib inline
pca_ss = PCA().fit(ss_crypto)
pca_mm = PCA().fit(mm_crypto)
pd.DataFrame([i for i in enumerate(np.cumsum(pca_ss.explained_variance_ratio_))]).hvplot.line(x= '0',y ='1',subplots =True)
pd.DataFrame([i for i in enumerate(np.cumsum(pca_mm.explained_variance_ratio_))]).hvplot.line(x= '0',y ='1',subplots =True)
pd_plot = pd.DataFrame([i for i in enumerate(np.cumsum(pca_ss.explained_variance_ratio_))]).merge(pd.DataFrame([i for i in enumerate(np.cumsum(pca_mm.explained_variance_ratio_))]),left_on = 0,right_on = 0)
pd_plot.columns = ['number of components','variance culmulated SS','variance culmulated mm']
pd_plot['number of components'] = pd_plot['number of components'].map(lambda x:x + 1)
pd_plot.hvplot(x = 'number of components',y = ['variance culmulated SS','variance culmulated mm'],subplots = True,width = 400)
```

```python
# in the plot above, for different scaler methods, standard scaler would needs 65 components to explain 80% variance, on the
# other hand maxmin method needs 15 to explain 80% of variance

def component_n(n):
    inertia = []
    for i in range(1,100):
        kmean = KMeans(i).fit(n)
        inertia_ = kmean.inertia_
        inertia.append(inertia_)
        df_inertia = pd.DataFrame({"k":range(0,80),'inertia':inertia})
    return df_inertia.plot(x = k,y = 'inertia')
component_n(pca_mm.predict(mm_crypto))
```
